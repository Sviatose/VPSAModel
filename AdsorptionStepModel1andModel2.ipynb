```python
# Initial data
# Initial parameters
# MODEL 1
#Constants
R = 8.314
# Gas flow
Flow_rate = 6.72701607326312E-07 # kmol-s
pressure = 3.2 # bar
# Bed parameters
L = 0.83  # m 
D = 0.021 #m

RHO = 715 #kg/m3
rho_p = 1060 #kg/m3
ei = 0.33 # External porosity
ep = 0.45 # Internal porosity
r_p = 9e-4 # Particle radius, m
d_pore = 800*10**(-9) # Macro pore diameter, m 
tortuosity = 2

# Thermal and kinetic parameters 
T_gas_in = 303 # K
T_solid_in = 303 # K

mu_g = 1.185e-2*0.001
k_gas = 2.198e-2
k_solid = 0
cp_gas = 34000 #J/kmol K
cp_solid = 878 #J/kgK

dH_ch4 = 33.674*10**6
dH_c2h6 = 0*10**6
dH_c3h8 = 0*10**6
dH_c4h10 = 0*10**6
dH_n2 = 0*10**6
dH_co2 = 38.947*10**6


# Simulation parameters
t_end =  3000 #seconds
dx = L/10

#Calculation of other parameters and initial composition 
c0 =  pressure * 100 / (R * T_gas_in) #kmol/m3 Molar concentration of gas
molar_fraction_ch4 = 0.55
molar_fraction_c2h6 = 0
molar_fraction_c3h8 = 0
molar_fraction_c4h10 = 0
molar_fraction_co2 = 0.45
molar_fraction_n2 = 0

c_ch4_initial = (molar_fraction_ch4) * c0  # Molar fraction of methane
c_c2h6_initial = (molar_fraction_c2h6) * c0 # Molar fraction of ethane
c_c3h8_initial = (molar_fraction_c3h8) * c0 # Molar fraction of propane
c_c4h10_initial = (molar_fraction_c4h10) * c0 # Molar fraction of butane
c_co2_initial = (molar_fraction_co2) * c0 # Molar fraction of CO2
c_n2_initial = (molar_fraction_n2) * c0 # Molar fraction of Nitrogen

Q_in = Flow_rate/c0
Q_out = Q_in
n = 1.05
Area = np.pi*D**2/4  # Cross-sectional area of bed
e = ei + ep # Total parameters
A_solid = (1-ei)*(3/r_p) # specific surface area, 1/m

u_initial = Q_in / (Area) # Velocity initial
u_last = Q_out / (Area)

x = np.arange(0,L+dx,dx) # Space matrice initialization


u_guess = [u_initial,u_last]
old_indices = np.arange(0,len(u_guess))
new_length = len(x)
new_indices = np.linspace(0,len(u_guess)-1,new_length)
spl = UnivariateSpline(old_indices,u_guess,k=1,s=0)
u = spl(new_indices)

# Isotherm parameters
IP1_CH4 = 1.079e-6 #kmol/kg
IP2_CH4 = 2208 #1/bar
IP3_CH4 = 0.916 # 1/bar
IP4_CH4 = -130.9 # K

  
IP1_C2H6 = 0.0027 #kmol/kg
IP2_C2H6 = 0.0 #1/bar 
IP3_C2H6 = 2.66E-04 # 1/bar
IP4_C2H6 = 2833.77 # K

IP1_C3H8 = 0.0062 #kmol/kg
IP2_C3H8 = 0.0 #1/bar 
IP3_C3H8 = 3.75E-04 # 1/bar
IP4_C3H8 = 2795.28 # K

IP1_C4H10 = 0.0076 #kmol/kg
IP2_C4H10 = 0.0 #1/bar 
IP3_C4H10 = 0.0015 # 1/bar
IP4_C4H10 = 2600 # K

IP1_CO2 = 1.127e-5 #kmol/kg
IP2_CO2 = 1880 #(kmol/kg)/bar 
IP3_CO2 = 0.145 # 1/bar
IP4_CO2 = 777.9 # K

IP1_N2 = 0 #kmol/kg
IP2_N2 = 0 #(kmol/kg)/bar 
IP3_N2 = 0 # 1/bar
IP4_N2 = 0 # K









def initialization(dx,u,n):

  x = np.arange(0,L+dx,dx) # Space matrice initialization 
  u_initial = Q_in / (Area) # Velocity initial
  u_last = 0 / (Area)

  u_guess = [u_initial,u_last]
  old_indices = np.arange(0,len(u_guess))
  new_length = len(x)
  new_indices = np.linspace(0,len(u_guess)-1,new_length)
  spl = UnivariateSpline(old_indices,u_guess,k=1,s=0)
  u = spl(new_indices)

  dt = e * (dx) /n*np.nanmax(u) #Time step estimation 
  t = np.arange(0,t_end+dt,dt) # Time calculation

  #Concentration matrice
  c_ch4 = np.zeros(len(x))
  c_c2h6 = np.zeros(len(x))
  c_c3h8 = np.zeros(len(x))
  c_c4h10 = np.zeros(len(x))
  c_co2 = np.zeros(len(x))
  c_n2 = np.zeros(len(x))

  #Initial parameters
  c_n2[:] = c0  # Bed initially filled with Nitrogen
  c_n2[0] = c_n2_initial
  c_co2[0] = c_co2_initial
  c_ch4[0] = c_ch4_initial
  c_c2h6[0] = c_c2h6_initial
  c_c3h8[0] = c_c3h8_initial
  c_c4h10[0] = c_c4h10_initial
  
  

  c_total = c_ch4 + c_c2h6 + c_c3h8+ c_c4h10 + c_n2 + c_co2

  #Thermal int.
  T_solid = np.zeros(len(x))
  T_solid[:] = T_solid_in 
  T_gas = np.zeros(len(x))
  T_gas[:] = T_gas_in 
  
  # Initial loading 
  w_ch4 = np.zeros(len(x))
  w_c2h6 = np.zeros(len(x))
  w_c3h8 = np.zeros(len(x))
  w_c4h10 = np.zeros(len(x))
  w_co2 = np.zeros(len(x))
  w_n2 = np.zeros(len(x))

  w_n2[:] = (IP1_N2*np.exp(IP2_N2/T_gas)) / (1 + IP3_N2*np.exp(IP4_N2/T_gas))

  #Matrices for right hand side of eq. 
  RHS_ch4 = np.zeros(len(x))
  RHS_c2h6 = np.zeros(len(x))
  RHS_c3h8 = np.zeros(len(x))
  RHS_c4h10 = np.zeros(len(x))
  RHS_co2 = np.zeros(len(x))
  RHS_n2 = np.zeros(len(x))

  RHS_solid_energy = np.zeros(len(x))
  RHS_gas_energy = np.zeros(len(x))
  
  #Equilibrium inital concentration
  weq_ch4 = np.zeros(len(x))
  weq_c2h6 = np.zeros(len(x))
  weq_c3h8 = np.zeros(len(x))
  weq_c4h10 = np.zeros(len(x))
  weq_co2 = np.zeros(len(x))
  weq_n2 = np.zeros(len(x))

  #Initial loading derivatives
  dwdt_ch4 = np.zeros(len(x))
  dwdt_c2h6 = np.zeros(len(x))
  dwdt_c3h8 = np.zeros(len(x))
  dwdt_c4h10 = np.zeros(len(x))
  dwdt_co2 = np.zeros(len(x))
  dwdt_n2 = np.zeros(len(x))

  #Initial adsorbed amount
  adsorbed_ch4 = np.zeros(len(x))
  adsorbed_c2h6 = np.zeros(len(x))
  adsorbed_c3h8 = np.zeros(len(x))
  adsorbed_c4h10 = np.zeros(len(x))
  adsorbed_co2 = np.zeros(len(x))
  adsorbed_n2 = np.zeros(len(x))

  #Initial velocity gradient
  dudx = np.zeros(len(x))
  #Time now 
  time = 0

  #For plotting: 
  c_ch4_graph = c_ch4.copy()
  c_c2h6_graph = c_c2h6.copy()
  c_c3h8_graph = c_c3h8.copy()
  c_c4h10_graph = c_c4h10.copy()
  c_co2_graph = c_co2.copy()
  c_n2_graph = c_n2.copy()
  u_graph = u.copy()
  T_solid_graph = T_solid.copy()
  T_gas_graph = T_gas.copy()

  t_graph = [time]


  print("INITIAL CONDITIONS ARE SET")
  return (x, dudx, c_total, dt, t,u, 
          c_ch4 , c_c2h6,c_c3h8,c_c4h10,c_co2, c_n2, c_ch4_graph,
          c_c2h6_graph,c_c3h8_graph,c_c4h10_graph,c_co2_graph,c_n2_graph,
          T_solid_graph,T_gas_graph,u_graph, 
          RHS_ch4, RHS_c2h6, RHS_n2,RHS_c3h8,RHS_c4h10,RHS_co2,
          RHS_solid_energy, RHS_gas_energy,
          weq_ch4, weq_c2h6, weq_c3h8,weq_c4h10, weq_co2 , weq_n2,
          dwdt_ch4, dwdt_c2h6,dwdt_c3h8,dwdt_c4h10,dwdt_co2, dwdt_n2,
          adsorbed_ch4, adsorbed_c2h6, adsorbed_c3h8,
          adsorbed_c4h10,adsorbed_co2, adsorbed_n2,
          w_ch4 , w_c2h6,w_c3h8,w_c4h10,w_co2, w_n2, time,
          t_graph, T_solid, T_gas)

def Equilibrium_data(T_solid):
  #Parameters for calculating Langmuir isotherm
  Qmax_CH4 = IP1_CH4*np.exp(IP2_CH4/T_solid)# kmol-kg
  b_CH4 = IP3_CH4*np.exp(IP4_CH4/T_solid)# 1-bar

  Qmax_C2H6 = IP1_C2H6 - IP2_C2H6*T_solid # kmol-kg
  b_C2H6 = IP3_C2H6*np.exp(IP4_C2H6/T_solid) # 1-bar

  Qmax_C3H8 = IP1_C3H8 - IP2_C3H8*T_solid # kmol-kg
  b_C3H8 = IP3_C3H8*np.exp(IP4_C3H8/T_solid) # 1-bar

  Qmax_C4H10 = IP1_C4H10 - IP2_C4H10*T_solid # kmol-kg
  b_C4H10 = IP3_C4H10*np.exp(IP4_C4H10/T_solid) # 1-bar

  Qmax_CO2 = IP1_CO2 *np.exp(IP2_CO2/T_solid)# kmol-kg
  b_CO2 = IP3_CO2*np.exp(IP4_CO2/T_solid) # 1-bar

  Qmax_N2 = IP1_N2*np.exp(IP2_N2/T_solid)# kmol-kg
  b_N2 = IP3_N2*np.exp(IP4_N2/T_solid) # 1-bar

  return (Qmax_CH4, b_CH4, Qmax_C2H6, b_C2H6, Qmax_C3H8, 
          b_C3H8,Qmax_C4H10, b_C4H10, Qmax_N2, b_N2, Qmax_CO2, b_CO2)
  



def equilibrium():

    #Function to calculate equilibrium point based on extended Langmuir 
    weq_ch4 = (((Qmax_CH4 * 1 * (c_ch4 / c_total) * pressure)) / 
               (1 + b_CH4 * (c_ch4 / c_total) * pressure + 
                    b_C2H6 *(c_c2h6 / c_total) * pressure+ 
                  + b_C3H8 * (c_c3h8 / c_total)* pressure+ 
                  + b_C4H10 * (c_c4h10 / c_total)* pressure+
                + 0*b_CO2* (c_co2 / c_total)* pressure+
                    b_N2 * (c_n2 / c_total)* pressure))
    
    weq_c2h6 = (((Qmax_C2H6 * b_C2H6 * (c_c2h6 / c_total) * pressure)) / 
              (1 + b_CH4 * (c_ch4 / c_total) * pressure  
                 + b_C2H6 * (c_c2h6 / c_total)* pressure+
                 + b_C3H8 * (c_c3h8 / c_total)* pressure+
                 + b_C4H10 * (c_c4h10 / c_total)* pressure+ 
                  + b_CO2* (c_co2 / c_total)* pressure+ 
                   b_N2 * (c_n2 / c_total)* pressure))
    
    weq_c3h8 = (((Qmax_C3H8 * b_C3H8 * (c_c3h8 / c_total) * pressure)) / 
                (1 + b_CH4 * (c_ch4 / c_total) * pressure  
                 + b_C2H6 * (c_c2h6 / c_total)* pressure+ 
                 + b_C3H8 * (c_c3h8 / c_total)* pressure+
                 + b_C4H10 * (c_c4h10 / c_total)* pressure+ 
                 + b_CO2* (c_co2 / c_total)* pressure+
                   b_N2 * (c_n2 / c_total)* pressure))

    weq_c4h10 = (((Qmax_C4H10 * b_C4H10 * (c_c4h10 / c_total) * pressure)) / 
                (1 + b_CH4 * (c_ch4 / c_total) * pressure  
                 + b_C2H6 * (c_c2h6 / c_total)* pressure+ 
                 + b_C3H8 * (c_c3h8 / c_total)* pressure+ 
                 + b_C4H10 * (c_c4h10 / c_total)* pressure+
                 + b_CO2* (c_co2 / c_total)* pressure+
                   b_N2 * (c_n2 / c_total)* pressure))
    
    weq_co2 = (((Qmax_CO2 * 1 * (c_co2 / c_total) * pressure)) / 
                (1 + 0*b_CH4 * (c_ch4 / c_total) * pressure  
                 + b_C2H6 * (c_c2h6 / c_total)* pressure+ 
                 + b_C3H8 * (c_c3h8 / c_total)* pressure+ 
                 + b_C4H10 * (c_c4h10 / c_total)* pressure+
                 + b_CO2* (c_co2 / c_total)* pressure+
                   b_N2 * (c_n2 / c_total)* pressure))
    
    weq_n2 = (((Qmax_N2 * 1 * (c_n2 / c_total) * pressure)) / 
                (1 + b_CH4 * (c_ch4 / c_total) * pressure  
                   + b_C2H6 * (c_c2h6 / c_total)* pressure+
                   + b_C3H8 * (c_c3h8 / c_total)* pressure+ 
                   + b_C4H10 * (c_c4h10 / c_total)* pressure+
                   + b_CO2* (c_co2 / c_total)* pressure+
                     b_N2 * (c_n2 / c_total)* pressure))
    
    return weq_ch4, weq_c2h6, weq_c3h8,weq_c4h10, weq_co2, weq_n2

def MassTransferCoefficient():
  M_ch4 = 16 # Molecular weight of methane
  M_c2h6 = 30 # Molecular weight of ethane
  M_c3h8 = 44 # Molecular weight of propane
  M_c4h10 = 58 #Molecular weight of butane
  M_N2 = 28 # Molecular weight of nitrogen
  M_CO2 = 44 # Molecular weight of carbon dioxide
  

  M = ((c_ch4_initial / (pressure*100/(R*T_gas_in))) * M_ch4 
       + (c_c2h6_initial / (pressure*100/(R*T_gas_in))) * M_c2h6 
       + (c_n2_initial / (pressure*100/(R*T_gas_in))) * M_N2
       + (c_c3h8_initial / (pressure*100/(R*T_gas_in))) * M_c3h8 
       + (c_c4h10_initial / (pressure*100/(R*T_gas_in))) * M_c4h10 
       + (c_co2_initial / (pressure*100/(R*T_gas_in))) * M_CO2)
  rho_gas = (pressure*100/(R*T_gas_in)) * M
  M_gas =  M # Moldecular weight of gas 

  Diff_vol_CO2 = 26.7 # Diffusion volume of CO2. Tabulated value 
  Diff_vol_C = 15.9 # Diffusion volume of carbon. Tabulated value 
  Diff_vol_H = 2.31 # Diffusion volume of hydrogen. Tabulated value 
  Diff_vol_N2 = 18.5 # Diffusion volume of nitrogen. Tabulated value 
  Diff_vol_gas = 19.7 # Diffusion volume of gas (air). Tabulated value 


  # Diffusion volume of methane based on the chemical formula CH4 
  Diff_vol_ch4 = 1*Diff_vol_C + 4 * Diff_vol_H 
  # Diffusion volume of ethane based on the chemical formula C2H6 
  Diff_vol_c2h6 = 2*Diff_vol_C + 6 * Diff_vol_H
  # Diffusion volume of propane based on the chemical formula C3H8 
  Diff_vol_c3h8 = 3*Diff_vol_C + 8 * Diff_vol_H
  # Diffusion volume of propane based on the chemical formula C4H10 
  Diff_vol_c4h10 = 4*Diff_vol_C + 10 * Diff_vol_H

  #The Fuller-Schettler-Giddings Equation
  Dch4 = (((10 ** -3) * T_gas_in ** 1.75 * ((1/(M_ch4)) + (1/(M_gas))) ** (1/2)/ 
         (pressure*(Diff_vol_ch4 **(1/3) + Diff_vol_gas**(1/3))**2))*10**(-4))
  Dc2h6 = (((10 ** -3) * T_gas_in ** 1.75 * ((1/(M_c2h6)) + (1/(M_gas))) ** (1/2)/ 
          (pressure*(Diff_vol_c2h6 **(1/3) + Diff_vol_gas**(1/3))**2))*10**(-4))
  Dc3h8 = (((10 ** -3) * T_gas_in ** 1.75 * ((1/(M_c3h8)) + (1/(M_gas))) ** (1/2)/ 
          (pressure*(Diff_vol_c3h8 **(1/3) + Diff_vol_gas**(1/3))**2))*10**(-4))
  Dc4h10 = (((10 ** -3) * T_gas_in ** 1.75 * ((1/(M_c4h10)) + (1/(M_gas))) ** (1/2)/ 
          (pressure*(Diff_vol_c4h10 **(1/3) + Diff_vol_gas**(1/3))**2))*10**(-4))
  Dco2 = (((10 ** -3) * T_gas_in ** 1.75 * ((1/(M_CO2)) + (1/(M_gas))) ** (1/2)/ 
          (pressure*(Diff_vol_CO2 **(1/3) + Diff_vol_gas**(1/3))**2))*10**(-4))
  Dn2 = (((10 ** -3) * T_gas_in ** 1.75 * ((1/(M_N2)) + (1/(M_gas))) ** (1/2)/ 
          (pressure*(Diff_vol_N2 **(1/3) + Diff_vol_gas**(1/3))**2))*10**(-4))
  print("------")

  
  Sc_ch4 = (mu_g)/(rho_gas*Dch4)
  Sc_c2h6 = (mu_g)/(rho_gas*Dc2h6)
  Sc_c3h8 = (mu_g)/(rho_gas*Dc3h8)
  Sc_c4h10 = (mu_g)/(rho_gas*Dc4h10)
  Sc_co2 = (mu_g)/(rho_gas*Dco2)
  Sc_n2 = (mu_g)/(rho_gas*Dn2)

  Re = (rho_gas*(u_initial/ep)*2*r_p)/(mu_g)

  Sh_ch4 = 2.0 + 1.1*(Sc_ch4**(1/3))*(Re**(0.6))
  Sh_c2h6 = 2.0 + 1.1*(Sc_c2h6**(1/3))*(Re**(0.6))
  Sh_c3h8 = 2.0 + 1.1*(Sc_c3h8**(1/3))*(Re**(0.6))
  Sh_c4h10 = 2.0 + 1.1*(Sc_c4h10**(1/3))*(Re**(0.6))
  Sh_co2 = 2.0 + 1.1*(Sc_co2**(1/3))*(Re**(0.6))
  Sh_n2 = 2.0 + 1.1*(Sc_n2**(1/3))*(Re**(0.6))
  

  k_c_ch4 = Sh_ch4 * Dch4 / (2*r_p) # m/s
  k_c_c2h6 = Sh_c2h6 * Dc2h6 / (2*r_p) # m/s
  k_c_c3h8 = Sh_c3h8 * Dc3h8 / (2*r_p) # m/s
  k_c_c4h10 = Sh_c4h10 * Dc4h10 / (2*r_p) # m/s
  k_c_co2 = Sh_co2 * Dco2 / (2*r_p)
  k_c_n2 = Sh_n2 * Dc2h6 / (2*r_p)


  
  ''' Calculation of the effective diffusivity'''
  Dk_ch4 = (d_pore/3)*((8*8314 * T_gas_in)/(np.pi*M_ch4))**(1/2)
  Dk_c2h6 = (d_pore/3)*((8*8314 * T_gas_in)/(np.pi*M_c2h6))**(1/2)
  Dk_c3h8 = (d_pore/3)*((8*8314 * T_gas_in)/(np.pi*M_c3h8))**(1/2)
  Dk_c4h10 = (d_pore/3)*((8*8314 * T_gas_in)/(np.pi*M_c4h10))**(1/2)
  Dk_co2 = (d_pore/3)*((8*8314 * T_gas_in)/(np.pi*M_CO2))**(1/2)
  Dk_n2 = (d_pore/3)*((8*8314 * T_gas_in)/(np.pi*M_N2))**(1/2)

  Deff_ch4 = (1)/(tortuosity)*(Dch4 * Dk_ch4)/(Dch4 +  Dk_ch4)
  Deff_c2h6 = (1)/(tortuosity)*(Dc2h6 * Dk_c2h6)/(Dc2h6 + Dk_c2h6)
  Deff_c3h8 = (1)/(tortuosity)*(Dc3h8 * Dk_c3h8)/(Dc3h8 + Dk_c3h8)
  Deff_c4h10 = (1)/(tortuosity)*(Dc4h10 * Dk_c4h10)/(Dc4h10 + Dk_c4h10)
  Deff_co2 = (1)/(tortuosity)*(Dco2 * Dk_co2)/(Dco2 + Dk_co2)
  Deff_n2 = (1)/(tortuosity)*(Dn2 * Dk_n2)/(Dn2 + Dk_n2)

  k_h_ch4 = (R/(10**5))*1000*T_gas_in*(RHO/(1-(ei+ep))/(ei))*((Qmax_CH4*b_CH4)) 
  k_h_c2h6 = (R/(10**5))*1000*T_gas_in*(RHO/(1-(ei+ep))/(ei))*((Qmax_C2H6*b_C2H6))
  k_h_c3h8 = (R/(10**5))*1000*T_gas_in*(RHO/(1-(ei+ep))/(ei))*((Qmax_C3H8*b_C3H8))
  k_h_c4h10 = (R/(10**5))*1000*T_gas_in*(RHO/(1-(ei+ep))/(ei))*((Qmax_C4H10*b_C4H10))
  k_h_co2 = (R/(10**5))*1000*T_gas_in*(RHO/(1-(ei+ep))/(ei))*((Qmax_CO2*b_CO2))       
  k_h_n2 = (R/(10**5))*1000*T_gas_in*(RHO/(1-(ei+ep))/(ei))*((Qmax_N2*b_N2))

  
  k_ch4 = (((r_p*k_h_ch4)/(3*k_c_ch4) 
  + ((r_p)**2*k_h_ch4)/(15*Deff_ch4*ep))**(-1))

  k_c2h6 = (((r_p*k_h_c2h6)/(3*k_c_c2h6) 
  + ((r_p)**2*k_h_c2h6)/(15*Deff_c2h6*ep))**(-1)) 

  k_c3h8 = (((r_p*k_h_c3h8)/(3*k_c_c3h8) 
  + ((r_p)**2*k_h_c3h8)/(15*Deff_c3h8*ep))**(-1))

  k_c4h10 = (((r_p*k_h_c4h10)/(3*k_c_c4h10) +
              ((r_p)**2*k_h_c4h10)/(15*Deff_c4h10*ep))**(-1))
  
  k_co2 = (((r_p*k_h_co2)/(3*k_c_co2)
   + ((r_p)**2*k_h_co2)/(15*Deff_co2*ep))**(-1))
  
  k_n2 = (((r_p*k_h_n2)/(3*k_c_n2) + ((r_p)**2*k_h_n2)/(15*Deff_n2*ep))**(-1))
  return [5.7E-5,0,0,0,0.025,0]

def adsorption_derivative():
    dwdt_ch4 = MTC[0] *(weq_ch4 - w_ch4) 
    dwdt_c2h6 = MTC[1] * (weq_c2h6 - w_c2h6) 
    dwdt_c3h8 = MTC[2] * (weq_c3h8 - w_c3h8)
    dwdt_c4h10 = MTC[3] * (weq_c4h10 - w_c4h10)
    dwdt_co2 = MTC[4] * (weq_co2 - w_co2)  
    dwdt_n2 = MTC[5] * (weq_n2 - w_n2)
    return dwdt_ch4, dwdt_c2h6, dwdt_c3h8,dwdt_c4h10,dwdt_co2,dwdt_n2

def time_integration_adsorption(dwdt_ch4,dwdt_c2h6,dwdt_c3h8,dwdt_c4h10,dwdt_co2,
                                dwdt_n2,w_ch4,w_c2h6,w_c3h8,w_c4h10,w_co2,w_n2):
                              
    w_ch4 = w_ch4 + dwdt_ch4 * dt
    w_c2h6 = w_c2h6 + dwdt_c2h6* dt
    w_c3h8 = w_c3h8 + dwdt_c3h8* dt
    w_c4h10 = w_c4h10 + dwdt_c4h10* dt
    w_co2 = w_co2 + dwdt_co2* dt
    w_n2 = w_n2 + dwdt_n2* dt

    return w_ch4, w_c2h6,w_c3h8,w_c4h10,w_co2,w_n2

def calculation_adsorbed_to_extract_from_mass():
  adsorbed_ch4 = RHO * dwdt_ch4 #kmol- m3
  adsorbed_c2h6 = RHO * dwdt_c2h6 #kmol - m3
  adsorbed_c3h8 = RHO * dwdt_c3h8 #kmol - m3
  adsorbed_c4h10 = RHO * dwdt_c4h10 #kmol - m3
  adsorbed_co2 = RHO * dwdt_co2
  adsorbed_n2 = RHO * dwdt_n2 
  return adsorbed_ch4, adsorbed_c2h6,adsorbed_c3h8,adsorbed_c4h10,adsorbed_co2, adsorbed_n2

def RHS_Mass_Transfer(): 

    dudx[1:] = ((-adsorbed_c4h10[1:]-adsorbed_c3h8[1:] - adsorbed_c2h6[1:] - 
                 adsorbed_ch4[1:] - adsorbed_co2[1:] 
                 -adsorbed_n2[1:])/c_total[1:])
     
    u[1:] = u[:-1] + dudx[1:] * dx

    RHS_ch4[1:] = (- (u[:-1]) / (e * dx) * (c_ch4[1:] - c_ch4[:-1]) - 
    adsorbed_ch4[1:]/e - (dudx[1:])*(c_ch4[1:])/e)

    RHS_c2h6[1:] = (- (u[:-1]) / (e * dx) * (c_c2h6[1:] - c_c2h6[:-1]) - 
                    adsorbed_c2h6[1:]/e - (dudx[1:])*(c_c2h6[1:])/e)
    
    RHS_c3h8[1:] = (- (u[:-1]) / (e * dx) * (c_c3h8[1:] - c_c3h8[:-1]) - 
                    adsorbed_c3h8[1:]/e - (dudx[1:])*(c_c3h8[1:])/e)
    
    RHS_c4h10[1:] = (- (u[:-1]) / (e * dx) * (c_c4h10[1:] - c_c4h10[:-1]) - 
                    adsorbed_c4h10[1:]/e - (dudx[1:])*(c_c4h10[1:])/e)
    
    RHS_co2[1:] = (- (u[:-1]) / (e * dx) * (c_co2[1:] - c_co2[:-1]) - 
                    adsorbed_co2[1:]/e - (dudx[1:])*(c_co2[1:])/e)
    
    RHS_n2[1:] = (- (u[:-1]) / (e * dx) * (c_n2[1:] - c_n2[:-1]) - 
                    adsorbed_n2[1:]/e - (dudx[1:])*(c_n2[1:])/e)
    
    return RHS_ch4, RHS_c2h6, RHS_c3h8, RHS_c4h10, RHS_co2, RHS_n2

def HTC():
  #Function to calculate heat transfer coefficient in the process
  M_ch4 = 16 # Molecular weight of methane
  M_c2h6 = 30 # Molecular weight of ethane
  M_c3h8 = 44 # Molecular weight of propane
  M_c4h10 = 58 #Molecular weight of butane
  M_N2 = 28 # Molecular weight of nitrogen
  M_CO2 = 44 # Molecular weight of carbon dioxide
  M_gas = 29 # Moldecular weight of gas (air)

  M = ((((c_ch4 / (pressure*100/(R*T_gas))) * M_ch4 + 
       (c_c2h6 / (pressure*100/(R*T_gas))) * M_c2h6 + 
       (c_c3h8 / (pressure*100/(R*T_gas))) * M_c3h8 + 
       (c_c4h10 / (pressure*100/(R*T_gas))) * M_c4h10 +
        c_co2 / (pressure*100/(R*T_gas))) * M_CO2 + 
       c_n2 / (pressure*100/(R*T_gas))) * M_N2)
  
  rho_gas = c_total * M
  Re = rho_gas*2*abs(u)*r_p / mu_g
  Pr = mu_g * cp_gas / (k_gas*M)
  if np.max(Re) <190:
    j = 1.66 * Re ** (-0.51)
  else:
    j = 0.983 * Re ** (-0.41)
  h_gas_solid = j * cp_gas * abs(u) * c_total * Pr ** (-2/3)
  return h_gas_solid 

def RHS_energy_f():

  RHS_solid_energy[0] = ((adsorbed_c4h10[0] * dH_c4h10 + 
   adsorbed_c3h8[0] * dH_c3h8 + adsorbed_c2h6[0] * dH_c2h6 
  + adsorbed_ch4[0] * dH_ch4 + adsorbed_co2[0] * dH_co2 + adsorbed_n2[0] * dH_n2) 
  / (RHO*cp_gas*(w_ch4[0]+w_c2h6[0]+w_c3h8[0]+w_c4h10[0]+w_co2[0]+w_n2[0])+RHO*cp_solid) 
  + h_gas_solid[0] * A_solid * (T_gas[0] - T_solid[0])
  / (RHO*cp_gas*(w_ch4[0]+w_c2h6[0] +w_c3h8[0]+w_c4h10[0]+w_co2[0] + w_n2[0]) + RHO*cp_solid) + 4*(hw/D)*(Tamb-T_solid[0])*10**6 + k_solid*(T_solid[1]-2*T_solid[0]+313)/(dx**2*(RHO*cp_gas*(w_ch4[0]+w_c2h6[0] +w_c3h8[0]+w_c4h10[0]+w_co2[0] + w_n2[0]) + RHO*cp_solid)))

  RHS_solid_energy[1:len(x)-1] = ((adsorbed_co2[1:len(x)-1] * dH_co2 + 
  adsorbed_c4h10[1:len(x)-1] * dH_c4h10  + adsorbed_c3h8[1:len(x)-1] * dH_c3h8  
  + adsorbed_c2h6[1:len(x)-1] * dH_c2h6  + adsorbed_ch4[1:len(x)-1] * dH_ch4  
  + adsorbed_n2[1:len(x)-1] * dH_n2 )
  /(RHO* cp_gas*(w_ch4[1:len(x)-1]+w_c2h6[1:len(x)-1]+w_c3h8[1:len(x)-1]+w_c4h10[1:len(x)-1]+w_co2[1:len(x)-1] 
  +w_n2[1:len(x)-1]) +
  RHO*cp_solid)+h_gas_solid[1:len(x)-1] * A_solid * (T_gas[1:len(x)-1] - T_solid[1:len(x)-1])
   / (e*cp_gas*(w_ch4[1:len(x)-1]+w_c2h6[1:len(x)-1]+
    w_c3h8[1:len(x)-1]+w_c4h10[1:len(x)-1] +w_co2[1:len(x)-1] +w_n2[1:len(x)-1]) + RHO*cp_solid) + 4*(hw/D)*(Tamb-T_solid[1:len(x)-1])*10**6 + k_solid*(T_solid[2:len(x)]-2*T_solid[1:len(x)-1]+T_solid[:len(x)-2])/(dx**2*(RHO*cp_gas*(w_ch4[1:len(x)-1]+w_c2h6[1:len(x)-1] +w_c3h8[1:len(x)-1]+w_c4h10[1:len(x)-1]+w_co2[1:len(x)-1] + w_n2[1:len(x)-1]) + RHO*cp_solid))) 

  RHS_solid_energy[len(x)-1] = ((adsorbed_c4h10[len(x)-1] * dH_c4h10 + 
  adsorbed_c3h8[len(x)-1] * dH_c3h8+ 
  adsorbed_c2h6[len(x)-1] * dH_c2h6 + adsorbed_ch4[len(x)-1] * dH_ch4 + 
  adsorbed_co2[len(x)-1] * dH_co2 +  adsorbed_n2[len(x)-1] * dH_n2)/ 
  (RHO* cp_gas*(w_ch4[len(x)-1]+ w_c2h6[len(x)-1]+ w_c3h8[len(x)-1] +
  + w_c4h10[len(x)-1]+ w_co2[len(x)-1] + w_n2[len(x)-1]) + RHO*cp_solid)
  + h_gas_solid[len(x)-1] * A_solid * (T_gas[len(x)-1] - T_solid[len(x)-1])/
  (RHO* cp_gas*(w_ch4[len(x)-1]+w_c2h6[len(x)-1]+
  w_c3h8[len(x)-1]+w_c4h10[len(x)-1] + w_n2[len(x)-1])+ w_co2[len(x)-1] + RHO*cp_solid) + 4*(hw/D)*(Tamb-T_solid[len(x)-1])*10**6 + k_solid*(-T_solid[len(x)-1]+T_solid[len(x)-2])/(dx**2*(RHO*cp_gas*(w_ch4[len(x)-1]+w_c2h6[len(x)-1] 
                                                                                                                                                       +w_c3h8[len(x)-1]+w_c4h10[len(x)-1]+w_co2[len(x)-1] + w_n2[len(x)-1]) + RHO*cp_solid)))

  RHS_gas_energy[len(x)-1] = (- (u[len(x)-2] / (e * dx))* (T_gas[len(x)-1] - 
  T_gas[len(x)-2]) + h_gas_solid[len(x)-1] * A_solid * (T_solid[len(x)-2] - 
  T_gas[len(x)-2]) /(e *cp_gas*c_total[len(x)-1]))
  
  RHS_gas_energy[1:] = (- (u[:-1] / (e * dx))* (T_gas[1:] - T_gas[:-1]) + 
  h_gas_solid[1:] * A_solid * (T_solid[1:] - T_gas[1:]) /(e *cp_gas*c_total[1:]))

  return RHS_solid_energy, RHS_gas_energy


print("STARTING CALCULATION TIME = 0")

(x, dudx, c_total, dt, t,u, 
          c_ch4 , c_c2h6,c_c3h8,c_c4h10,c_co2, c_n2, c_ch4_graph,
          c_c2h6_graph,c_c3h8_graph,c_c4h10_graph,c_co2_graph,c_n2_graph,
          T_solid_graph,T_gas_graph,u_graph, 
          RHS_ch4, RHS_c2h6, RHS_n2,RHS_c3h8,RHS_c4h10,RHS_co2,
          RHS_solid_energy, RHS_gas_energy,
          weq_ch4, weq_c2h6, weq_c3h8,weq_c4h10, weq_co2 , weq_n2,
          dwdt_ch4, dwdt_c2h6,dwdt_c3h8,dwdt_c4h10,dwdt_co2, dwdt_n2,
          adsorbed_ch4, adsorbed_c2h6, adsorbed_c3h8,
          adsorbed_c4h10,adsorbed_co2, adsorbed_n2,
          w_ch4 , w_c2h6,w_c3h8,w_c4h10,w_co2, w_n2, time,
          t_graph, T_solid, T_gas) = initialization(dx,u,n) 

(Qmax_CH4, b_CH4, Qmax_C2H6, b_C2H6, 
 Qmax_C3H8, b_C3H8,Qmax_C4H10, 
 b_C4H10,Qmax_N2, b_N2, Qmax_CO2, b_CO2)  = Equilibrium_data(T_solid_in)
 
MTC = MassTransferCoefficient()
print("Estimated Mass Transfer Coefficients")
print(MTC)
time_prev = int(time)

pbar = tqdm(total = time)
while t_end > time:
  u_prev = u
  c_total = c_ch4 + c_c2h6+ c_c3h8 + c_c4h10+ c_co2 + c_n2
  (c_ch4_prev, c_c2h6_prev,c_c3h8_prev,
   c_c4h10_prev ,c_co2_prev,c_n2_prev) = (c_ch4, c_c2h6,
  c_c3h8,c_c4h10,c_co2, c_n2)

  T_solid_prev = T_solid
  T_gas_prev = T_gas

  (Qmax_CH4, b_CH4, Qmax_C2H6, b_C2H6, 
 Qmax_C3H8, b_C3H8,Qmax_C4H10, 
 b_C4H10,Qmax_N2, b_N2, Qmax_CO2, b_CO2) = Equilibrium_data(T_solid)

  weq_ch4, weq_c2h6,weq_c3h8,weq_c4h10,weq_co2, weq_n2 = equilibrium()
  (dwdt_ch4, dwdt_c2h6,dwdt_c3h8,
   dwdt_c4h10,dwdt_co2,dwdt_n2) = adsorption_derivative()

  (adsorbed_ch4, adsorbed_c2h6,adsorbed_c3h8,
   adsorbed_c4h10,
   adsorbed_co2,adsorbed_n2) = calculation_adsorbed_to_extract_from_mass()
  
  RHS_ch4, RHS_c2h6,RHS_c3h8,RHS_c4h10,RHS_co2,RHS_n2 = RHS_Mass_Transfer()
  h_gas_solid = HTC()

  #RHS_solid_energy, RHS_gas_energy = RHS_energy_f() 
  w_ch4, w_c2h6,w_c3h8,w_c4h10,w_co2,w_n2 = time_integration_adsorption(dwdt_ch4,
  dwdt_c2h6, dwdt_c3h8,dwdt_c4h10,dwdt_co2,dwdt_n2,w_ch4,
  w_c2h6,w_c3h8,w_c4h10,w_co2,w_n2)

  c_ch4 = c_ch4_prev + RHS_ch4*dt
  c_c2h6 = c_c2h6_prev + RHS_c2h6*dt
  c_c3h8 = c_c3h8_prev + RHS_c3h8*dt
  c_c4h10 = c_c4h10_prev + RHS_c4h10*dt
  c_co2 = c_co2_prev + RHS_co2*dt
  c_n2 = c_n2_prev + RHS_n2*dt

  #T_solid = T_solid_prev + RHS_solid_energy*dt
  #T_gas = T_gas_prev + RHS_gas_energy*dt

  if int(time) != time_prev:
    pbar.update(1)

    c_ch4_graph = np.vstack((c_ch4_graph,c_ch4))
    c_c2h6_graph = np.vstack((c_c2h6_graph,c_c2h6))
    c_c3h8_graph = np.vstack((c_c3h8_graph,c_c3h8))
    c_c4h10_graph = np.vstack((c_c4h10_graph,c_c4h10))
    c_co2_graph = np.vstack((c_co2_graph,c_co2))
    c_n2_graph = np.vstack((c_n2_graph,c_n2))
    u_graph = np.vstack((u_graph,u))
    T_solid_graph = np.vstack((T_solid_graph,T_solid))
    T_gas_graph = np.vstack((T_gas_graph,T_gas))
    t_graph.append(time)
    
  time_prev = int(time)
  unstable = (np.any(c_ch4<0)) or (np.any(c_c2h6<0)) or (np.any(np.isnan(c_ch4))) or (np.any(T_gas<0))
  if not unstable:
    time += dt
  if unstable:
    n = 1.2*n
    time = 0
    print("Unstable, reducing time step")
    dx = dx/1.5
    (x, dudx, c_total, dt, t,u, 
          c_ch4 , c_c2h6,c_c3h8,c_c4h10,c_co2, c_n2, c_ch4_graph,
          c_c2h6_graph,c_c3h8_graph,c_c4h10_graph,c_co2_graph,c_n2_graph,
          T_solid_graph,T_gas_graph,u_graph, 
          RHS_ch4, RHS_c2h6, RHS_n2,RHS_c3h8,RHS_c4h10,RHS_co2,
          RHS_solid_energy, RHS_gas_energy,
          weq_ch4, weq_c2h6, weq_c3h8,weq_c4h10, weq_co2 , weq_n2,
          dwdt_ch4, dwdt_c2h6,dwdt_c3h8,dwdt_c4h10,dwdt_co2, dwdt_n2,
          adsorbed_ch4, adsorbed_c2h6, adsorbed_c3h8,
          adsorbed_c4h10,adsorbed_co2, adsorbed_n2,
          w_ch4 , w_c2h6,w_c3h8,w_c4h10,w_co2, w_n2, time,
          t_graph, T_solid, T_gas) = initialization(dx,u_prev,n) 


print("Finished successfully ")

fig1 = plt.figure()
ax1 = fig1.add_subplot(111)

major_ticks_x= np.arange(0, t_end+0.1, 100)
minor_ticks_x = np.arange(0, t_end+0.1, 20)

ax1.set_xticks(major_ticks_x)
ax1.set_xticks(minor_ticks_x, minor=True)


plt.rcParams['figure.figsize'] = [12, 12]
plt.rcParams.update({'font.size' : 18})
ax1.plot(t_graph, c_ch4_graph[:,-1]/c0, label = "Methane")
ax1.plot(t_graph, c_c2h6_graph[:,-1]/c0,  label = "Ethane")
ax1.plot(t_graph, c_c3h8_graph[:,-1]/c0,  label = "Propane")
ax1.plot(t_graph, c_c4h10_graph[:,-1]/c0,  label = "Butane")
ax1.plot(t_graph, c_co2_graph[:,-1]/c0,  label = "Carbon Dioxide")
ax1.plot(t_graph, c_n2_graph[:,-1]/c0,  label = "Nitrogen")
ax1.grid(which='both')
ax1.legend()
fig2 = plt.figure()
ax2 = fig2.add_subplot(111)
ax2.set_xticks(major_ticks_x)
ax2.set_xticks(minor_ticks_x, minor=True)


plt.rcParams['figure.figsize'] = [12, 12]
plt.rcParams.update({'font.size' : 18})
ax2.plot(t_graph, T_gas_graph[:,-1],label="solid", c = "b")
ax2.plot(t_graph, T_solid_graph[:,-1],label="solid", c = "b")
ax2.grid()

plt.show()

```

```python
#Model 3
seg = 1000
print("Staarted")
MTC = [5.7E-5,0.025]
m = GEKKO(remote = False)  # create GEKKO model
tf =  3000
nt = int(tf/10) + 1
m.time = np.linspace(0,tf,nt)
y1_feed = 0.55
y2_feed = 0.45
y3_feed = 0
L_seg = L/seg
Area = np.pi * D**2/4
u_initial = Flow_rate*R*T_initial/(Area*pressure*10**2)
P = [m.Var(3.2) for i in range(seg)] 
y1 =  [m.Var(0) for i in range(seg)] 
y2 =  [m.Var(0) for i in range(seg)]
y3 =  [m.Var(1) for i in range(seg)]
T  =  [m.Var(303) for i in range(seg)]
v =  [m.Var(0) for i in range(seg)]
q1 = [m.Var(0) for i in range(seg)]
q2 = [m.Var(0) for i in range(seg)]

# first segment Pressure 1
m.Equation(P[0]*y1[0].dt()+y1[0]*P[0].dt() == - (u_initial/(et))*(y1[0]*(P[0]-pressure)/(L_seg)+P[0]*(y1[0]-y1_feed)/(L_seg)) 
           - (P[0]*y1[0]/(et))*((v[0]-u_initial)/(L_seg)) - (rho/(10**2*et))*R*T[0]*q1[0].dt())
m.Equation([P[i]*y1[i].dt()+y1[i]*P[i].dt() == - (v[i]/(et))*(y1[i]*(P[i]-P[i-1])/(L_seg)+P[i]*(y1[i]-y1[i-1])/(L_seg)) 
           - (P[i]*y1[i]/(et))*((v[i]-v[i-1])/(L_seg)) - (rho/(10**2*et))*R*T[i]*q1[i].dt() for i in range(1,seg-1)])
m.Equation(P[seg-1]*y1[seg-1].dt()+y1[seg-1]*P[seg-1].dt() == - (v[seg-1]/(et))*(y1[seg-1]*(P[seg-1]-P[seg-2])/(L_seg)+P[seg-1]*(y1[seg-1]-y1[seg-2])/(L_seg)) 
           - (P[seg-1]*y1[seg-1]/(et))*((v[seg-1]-v[seg-2])/(L_seg)) - (rho/(10**2*et))*R*T[seg-1]*q1[seg-1].dt())


m.Equation(P[0]*y2[0].dt()+y2[0]*P[0].dt() == - (u_initial/(et))*(y2[0]*(P[0]-pressure)/(L_seg)+P[0]*(y2[0]-y2_feed)/(L_seg)) 
           - (P[0]*y2[0]/(et))*((v[0]-u_initial)/(L_seg)) - (rho/(10**2*et))*R*T[0]*q2[0].dt())

m.Equation([P[i]*y2[i].dt()+y2[i]*P[i].dt() == - (v[i]/(et))*(y2[i]*(P[i]-P[i-1])/(L_seg)+P[i]*(y2[i]-y2[i-1])/(L_seg)) 
           - (P[i]*y2[i]/(et))*((v[i]-v[i-1])/(L_seg)) - (rho/(10**2*et))*R*T[i]*q2[i].dt() for i in range(1,seg-1)])

m.Equation(P[seg-1]*y2[seg-1].dt()+y2[seg-1]*P[seg-1].dt() == - (v[seg-1]/(et))*(y2[seg-1]*(P[seg-1]-P[seg-2])/(L_seg)
            +P[seg-1]*(y2[seg-1]-y2[seg-2])/(L_seg)) 
           - (P[seg-1]**y2[seg-1]/(et))*((v[seg-1]-v[seg-2])/(L_seg)) - (rho/(10**2*et))*R*T[seg-1]*q2[seg-1].dt())

m.Equation(P[0]*y3[0].dt()+y3[0]*P[0].dt() == - (u_initial/(et))*(y3[0]*(P[0]-pressure)/(L_seg)+P[0]*(y3[0]-y3_feed)/(L_seg)) 
           - (P[0]*y3[0]/(et))*((v[0]-u_initial)/(L_seg)))
m.Equation([P[i]*y3[i].dt()+y3[i]*P[i].dt() == - (v[i]/(et))*(y3[i]*(P[i]-P[i-1])/(L_seg)+P[i]*(y3[i]-y3[i-1])/(L_seg)) 
           - (P[i]*y3[i]/(et))*((v[i]-v[i-1])/(L_seg)) for i in range(1,seg-1)])
m.Equation(P[seg-1]*y3[seg-1].dt()+y3[seg-1]*P[seg-1].dt() == - (v[seg-1]/(et))*(y3[seg-1]*(P[seg-1]-P[seg-2])/(L_seg)
                                                                                  +P[seg-1]*y3[seg-1]*(y3[seg-1]-y3[seg-2])/(L_seg)) 
           - (P[seg-1]*y3[seg-1]/(et))*((v[seg-1]-v[seg-2])/(L_seg)))


m.Equation((cp_gas*(P[0]*et*10**2/(R*T[0])) + cp_solid*rho)*T[0].dt() == 
            -(cp_gas*(P[0]*10**2/(R*T[0])))*u_initial*(T[0]-T_initial)/(L_seg) 
 - (rho*dH1*q1[0].dt() + rho*dH2*q2[0].dt())
            + 4*(hw/D)*(Tamb-T[0])*10**6)

m.Equation([(cp_gas*(P[i]*et*10**2/(R*T[i])) + cp_solid*rho)*T[i].dt() == 
            -(cp_gas*(P[i]*10**2/(R*T[i])))*v[i]*(T[i]-T[i-1])/(L_seg) 
 - (rho*dH1*q1[i].dt() + rho*dH2*q2[i].dt()) 
             + 4*(hw/D)*(Tamb-T[i])*10**6 for i in range(1,seg-1)])

m.Equation((cp_gas*(P[seg-1]*et*10**2/(R*T[seg-1])) + cp_solid*rho)*T[seg-1].dt() == 
            -(cp_gas*(P[seg-1]*10**2/(R*T[seg-1])))*v[seg-1]*(T[seg-1]-T[seg-2])/(L_seg) 
 - (rho*dH1*q1[seg-1].dt() + rho*dH2*q2[seg-1].dt()) 
             + 4*(hw/D)*(Tamb-T[seg-1])*10**6)


m.Equation(v[0] == -(1/K)*(P[0]-pressure)/(L_seg))
m.Equation([v[i] == -(1/K)*(P[i]-P[i-1])/(L_seg) for i in range(1,seg-1)])
m.Equation(v[seg-1] == -(1/K)*(P[seg-1]-P[seg-2])/(L_seg))

m.Equation([y1[i] +  y2[i]  == 1 - y3[i] for i in range(0,seg)])

m.Equation([q1[i].dt() == -MTC[0]*(q1[i] - (IP1_CH4*y1[i]*P[i]*exp**(IP2_CH4/T[i])/(1+y1[i]*P[i]*IP3_CH4*exp**(IP4_CH4/T[i])))) for i in range(0,seg)])
m.Equation([q2[i].dt() == -MTC[1]*(q2[i] - (IP1_CO2*y2[i]*P[i]*exp**(IP2_CO2/T[i])/(1+y2[i]*P[i]*IP3_CO2*exp**(IP4_CO2/T[i])))) for i in range(0,seg)])

# simulation
m.options.IMODE = 7
m.solve(disp = False)
print("FINIIIISHED")
```
